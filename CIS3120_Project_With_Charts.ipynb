```python
active_players = players.get_active_players()
```

```python
active_players_df = pd.DataFrame(active_players)
```

```python
active_players_df
```

```python
# @title Distribution of Player IDs

import matplotlib.pyplot as plt

# Create the histogram
plt.figure(figsize=(8, 6))
plt.hist(active_players_df['id'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Player ID')
plt.ylabel('Frequency')
_ = plt.title('Distribution of Player IDs')
```

```python
active_players_df.head(100)
```

```python
# @title Frequency of Shared First Names among Active Players

import matplotlib.pyplot as plt

first_name_counts = active_players_df['first_name'].value_counts()
plt.figure(figsize=(8, 6))
plt.hist(first_name_counts, bins=20, color='gold', edgecolor='black')
plt.xlabel('Number of Players Sharing the Same First Name')
plt.ylabel('Frequency')
_ = plt.title('Frequency of Shared First Names among Active Players')
```
